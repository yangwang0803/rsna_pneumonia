```python
import os
import pandas as pd
import numpy as np
from glob import glob

import pydicom

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import seaborn as sns
```

```python
# data
data_raw = pd.read_csv(r'C:\Users\3594997\Study\Kaggle\Pneumonia\stage\1\work_1.csv')
print(len(data_raw))
data_raw.head()
```

```python
# dir
dir_train = r'C:\Users\3594997\Study\Kaggle\Pneumonia\data\stage_1_train_images'
dir_test = r'C:\Users\3594997\Study\Kaggle\Pneumonia\data\stage_1_test_images'
```

```python
df = pd.DataFrame({'path': glob(os.path.join(dir_train, '*.dcm'))})
df['patientId'] = df['path'].map(lambda x: os.path.splitext(os.path.basename(x))[0])
print(len(df))
df.head()
```

```python
DCM_TAG_LIST = \
['AccessionNumber',
 'BitsAllocated',
 'BitsStored',
 'BodyPartExamined',
 'Columns',
 'ConversionType',
 'HighBit',
 'InstanceNumber',
 'LossyImageCompression',
 'LossyImageCompressionMethod',
 'Modality',
 'PatientAge',
 'PatientBirthDate',
 'PatientID',
 'PatientName',
 'PatientOrientation',
 'PatientSex',
 'PhotometricInterpretation',
 'PixelRepresentation',
 'PixelSpacing',
 'ReferringPhysicianName',
 'Rows',
 'SOPClassUID',
 'SOPInstanceUID',
 'SamplesPerPixel',
 'SeriesDescription',
 'SeriesInstanceUID',
 'SeriesNumber',
 'SpecificCharacterSet',
 'StudyDate',
 'StudyID',
 'StudyInstanceUID',
 'StudyTime',
 'ViewPosition']
def get_tags(in_path):
    c_dicom = pydicom.read_file(in_path, stop_before_pixels=True)
    tag_dict = {c_tag: getattr(c_dicom, c_tag, '') 
         for c_tag in DCM_TAG_LIST}
    tag_dict['path'] = in_path
    return pd.Series(tag_dict)
image_meta_df = df.apply(lambda x: get_tags(x['path']), 1)
# show the summary
image_meta_df['PatientAge'] = image_meta_df['PatientAge'].map(int)
```

```python
len(image_meta_df)
image_meta_df
```

```python
df = df.merge(image_meta_df, on='path', how='inner')
print(len(df))
df.head()
```

```python
data_raw = data_raw.merge(df, on='patientId', how='inner')
print(len(data_raw))
data_raw.head()
```

```python
data_raw.drop(columns='BodyPartExamined', inplace=True)
data_raw.head()
```

```python
data_raw = data_raw.assign(NumOfBox = data_raw.groupby('patientId').path.transform(lambda x: len(x)))
data_raw.head(10)
```
